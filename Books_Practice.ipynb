```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in  text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
books3['tags'] = books3['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(books3['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names() #our corpus words
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors) #similarity matrix
```

```python
similarity[0]
```

```python
def recommend(book):
    book_index = books3[books3['title']== book].index[0]
    distances = similarity[book_index]
    book_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x: x[1])[1:6]
    for i in book_list:
        print(books3.iloc[i[0]].title)
    
```
